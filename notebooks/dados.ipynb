```python
import pandas as pd
from IPython.display import display

# Definir o caminho do arquivo
file_path = r'C:\Users\Jesuino\Downloads\teste-25\data\nafld1.csv'

# Carregar o conjunto de dados
df = pd.read_csv(file_path)

# Exibir as primeiras linhas do DataFrame
print("Primeiras linhas do dataset:")
display(df.head(10))
```

```python
# Informações básicas sobre o conjunto de dados
print("Informações básicas sobre o dataset:")
info = pd.DataFrame({
    "Coluna": df.columns,
    "Tipo de Dado": df.dtypes,
    "Valores Não Nulos": df.count(),
    "Valores Ausentes": df.isnull().sum()
})
display(info)
```

```python
# Estatísticas descritivas para colunas numéricas
print("\nEstatísticas descritivas para colunas numéricas:")
display(df.describe())
```

```python
# Verificar valores ausentes
print("\nValores ausentes no dataset:")
missing_values = df.isnull().sum()
display(missing_values)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Box plots para variáveis numéricas
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, x=col)
    plt.title(f'Box plot de {col}')
    plt.show()

# Violino plots para variáveis numéricas
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.violinplot(data=df, x=col)
    plt.title(f'Violino plot de {col}')
    plt.show()
```

```python
# Verificar se há colunas de tipo "object" que deveriam ser convertidas
print("Verificação de tipos de dados errados:")
for col in df.select_dtypes(include=['object']).columns:
    unique_values = df[col].nunique()
    print(f"Coluna '{col}': {unique_values} valores únicos")
```
