```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o conjunto de dados
df = pd.read_csv('../data/train_data.csv')

# Visualizar as primeiras linhas do conjunto de dados
df.head()

# Histograma da Idade (age)
plt.figure(figsize=(6, 3.75))
sns.histplot(df['age'], bins=30, kde=True, color=sns.color_palette("viridis", as_cmap=True)(0.5))
plt.title('Histograma da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.tight_layout()
plt.savefig('histograma_idade.png')
plt.show()

# Gráfico de Dispersão: Peso (weight) vs. Altura (height)
plt.figure(figsize=(6, 3.75))
sns.scatterplot(data=df, x='weight', y='height', color=sns.color_palette("viridis", as_cmap=True)(0.7))
plt.title('Gráfico de Dispersão: Peso vs. Altura')
plt.xlabel('Peso')
plt.ylabel('Altura')
plt.tight_layout()
plt.savefig('grafico_dispersao_peso_altura.png')
plt.show()

# Gráfico de Barras: Status (status)
plt.figure(figsize=(6, 3.75))
sns.countplot(data=df, x='status', palette=sns.color_palette("viridis"))
plt.title('Gráfico de Barras: Status Vital')
plt.xlabel('Status Vital')
plt.ylabel('Contagem')
plt.tight_layout()
plt.savefig('grafico_barras_status.png')
plt.show()
```
