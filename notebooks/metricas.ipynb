```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import mlflow
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar a paleta de cores viridis
sns.set_palette("viridis")

# Carregar os dados
train_data = pd.read_csv('../data/train_data.csv')
test_data = pd.read_csv('../data/test_data.csv')

# Separar os dados em features e target
train_target = train_data['status']  # Troque aqui se o nome da coluna for diferente
test_target = test_data['status']  # Troque aqui se o nome da coluna for diferente
train_features = train_data.drop(columns=['status'])  # Troque aqui se o nome da coluna for diferente
test_features = test_data.drop(columns=['status'])  # Troque aqui se o nome da coluna for diferente

# Iniciar uma nova run no MLflow
mlflow.start_run()

# Treinar o modelo
model = LogisticRegression(solver='liblinear', random_state=42)
model.fit(train_features, train_target)

# Fazer previsões
y_pred = model.predict(test_features)
accuracy = accuracy_score(test_target, y_pred)
precision = precision_score(test_target, y_pred, average='weighted', zero_division=0)
recall = recall_score(test_target, y_pred, average='weighted', zero_division=0)
f1 = f1_score(test_target, y_pred, average='weighted', zero_division=0)
auc = roc_auc_score(test_target, model.predict_proba(test_features)[:, 1])

# Logar métricas no MLflow
mlflow.log_metric("accuracy", accuracy)
mlflow.log_metric("precision", precision)
mlflow.log_metric("recall", recall)
mlflow.log_metric("f1_score", f1)
mlflow.log_metric("auc_roc", auc)

# Salvar o modelo no MLflow
mlflow.sklearn.log_model(model, "model")
mlflow.end_run()

# Imprimir métricas no notebook
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"AUC-ROC: {auc:.4f}")

# Criar gráficos
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC-ROC']
scores = [accuracy, precision, recall, f1, auc]

plt.figure(figsize=(8, 4))
plt.barh(metrics, scores, color=sns.color_palette("viridis", len(metrics)))
plt.xlabel('Score')
plt.title('Model Performance Metrics')
plt.xlim(0, 1)
plt.show()

print("Modelo treinado e métricas registradas com MLflow!")
```
